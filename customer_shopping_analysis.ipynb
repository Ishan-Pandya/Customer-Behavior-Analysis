```python
# check the total number of null values in each column
df.isnull().sum()
```

```python
# we have 37 null values in review rating, we will replace the null values with the median value for the specific category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# null values has been replaced with median value of specific category
df.isnull().sum()
```

```python
# rename the column name in snakecase 
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.info()
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.age.min()
```

```python
df.age.max()
```

```python
# create new column age_group
age_lable = ['Yound Adult','Adult','Middle Aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = age_lable)
```

```python
# check the range for each age group
pd.qcut(df['age'], q=4)
```

```python
df[['age','age_group']].head(12)
```

```python
# check the distinct value in frewuency_of_purchase column as we will convert it into numerical value
df['frequency_of_purchases'].value_counts()
```

```python
# create new dictionary
frequency_mapping = {
    'Annually': 365,
    'Quarterly':90,
    'Every 3 Month':90,
    'Monthly' :30,
    'Bi-Weekly':14,
    'Fortnightly':14,
    'Weekly':7
}
```

```python
# create new column purchase_frequency_days
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(12)
```

```python
df[['promo_code_used','discount_applied']].head(12)
```

```python
# check if both the columns are identical
df['promo_code_used'].equals(df['discount_applied'])
```

```python
# drop promo_code_used column
df = df.drop('promo_code_used', axis=1)
```
